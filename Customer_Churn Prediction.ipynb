function will print all coloumn with unique values

```python
def print_unique_col_values(df):
       for column in df:
            if df[column].dtypes=='object':
                print(f'{column}: {df[column].unique()}') 
```

```python
print_unique_col_values(df1)
```

```python
df1.replace('No internet service','No',inplace=True) # chaning big statement to just no
df1.replace('No phone service','No',inplace=True)
```

```python
print_unique_col_values(df1)
```

No we know which column are with unique values and then we will assign 0 and 1 for no and yes

```python
yes_no_columns = ['Partner','Dependents','PhoneService','MultipleLines','OnlineSecurity','OnlineBackup',
                  'DeviceProtection','TechSupport','StreamingTV','StreamingMovies','PaperlessBilling','Churn']
for col in yes_no_columns:
    df1[col].replace({'Yes': 1,'No': 0},inplace=True)
```

```python
for col in df1:
    print(f'{col}: {df1[col].unique()}')
```

```python
df1['gender'].replace({'Female':1,'Male':0},inplace=True)
```

```python
df1.gender.unique() # gender is also converted into binary 
```

```python
df2 = pd.get_dummies(data=df1, columns=['InternetService','Contract','PaymentMethod'])
df2.columns
```

dummie will make seperated coloumns for different types of entries. Example: Internet services coloumn is now converted to 3 different coloumns like; DSL, Fibre optics and no

```python
df2.sample(5)
```

Finally! we have all values in numeric form

```python
df2.dtypes 
```

```python
cols_to_scale = ['tenure','MonthlyCharges','TotalCharges']

from sklearn.preprocessing import MinMaxScaler  
scaler = MinMaxScaler() # min max scalar set the values between 0 and 1
df2[cols_to_scale] = scaler.fit_transform(df2[cols_to_scale])
```

tenure, monthly charges and total chares have values between 0 and 1

```python
for col in df2:
    print(f'{col}: {df2[col].unique()}')
```

Train and testin splitit

```python
X = df2.drop('Churn',axis='columns')
y = df2['Churn']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=5)
```

80% is used for training and 20% for testing

```python
X_train.shape
```

```python
X_test.shape
```

```python
X_train[:10]
```

```python
len(X_train.columns)
```

ANN implementation

```python
import tensorflow as tf
from tensorflow import keras


model = keras.Sequential([
    keras.layers.Dense(26, input_shape=(26,), activation='relu'), # inut layer
    # Relu will output the input directly if it is positive, otherwise, it will output zero.
    keras.layers.Dense(15, activation='relu'), # hidden layer
    keras.layers.Dense(1, activation='sigmoid') # output layer
    # sigmoid guarantee that the output of this unit will always be between 0 and 1
])
  # Adam is an optimization algorithm that can be used instead of the classical stochastic gradient descent
  # procedure to update network weights iterative based in training data.
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

model.fit(X_train, y_train, epochs=100)
```

```python
model.evaluate(X_test, y_test)
```

```python
yp = model.predict(X_test)
yp[:5]
```

```python
y_pred = []
for element in yp:
    if element > 0.5:
        y_pred.append(1)
    else:
        y_pred.append(0)
```

```python
y_pred[:10]
```

```python
from sklearn.metrics import confusion_matrix , classification_report

print(classification_report(y_test,y_pred))
```
